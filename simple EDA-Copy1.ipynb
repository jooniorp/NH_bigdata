시계열데이터이므로 주식별로 lineplot 시각화

```python
tck_cd = list(stk_dd_info['tck_iem_cd'].unique())
len(tck_cd)
```

```python
stk_dd_info['tck_iem_cd'].value_counts()
```

```python
#크기 차이가 많이 나서 한 격자에 그리기엔 어려울 듯
for i in tck_cd[1:3]:
    sns.lineplot(data=stk_dd_info[stk_dd_info['tck_iem_cd']==i], x="bse_dt", y="tot_hld_act_cnt", label=i)
    plt.xticks(rotation=45)
plt.legend()
```

 주식마다 스케일이 다르므로 변동 편차를 확인해서 비교하려면 스케일링 후 편차를 고려해야 할 듯<br>
 1000개에서 990개 줄어든거랑 100개에서 90개로 줄어든 거랑 그 경향이 다르기 때문<br>
 보유 수량보단 보유비중비율만 사용해도 될 듯

```python
#tot_hld_act_cnt 시계열 그래프
fig, axes = plt.subplots(nrows=10, ncols=3, figsize=(30, 100))

for row in range(10):
    for col in range(3):
        idx = row * 5 + col
        ax = axes[row][col]
        ax.set_title(tck_cd[idx])
        sns.lineplot(data=stk_dd_info[stk_dd_info['tck_iem_cd']==tck_cd[idx]], x="bse_dt", y="tot_hld_act_cnt", ax=ax)
        plt.xticks(rotation=45)
```
